```python
# Function to create DataFrame from a table
def create_dataframe_from_table(connection, table_name):
    # Create a cursor
    cursor = connection.cursor()
    # Execute a query to select data from the table
    query = f"SELECT * FROM {table_name}"
    cursor.execute(query)
    # Fetch all rows from the result set
    data = cursor.fetchall()
    # Get the column names from the cursor description
    columns = [column[0] for column in cursor.description]
    # Create a DataFrame
    df = pd.DataFrame(data, columns=columns)
    # Close the cursor
    cursor.close()
    return df
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Function to create a DataFrame from a database query result
def create_dataframe_from_query(cursor, query):
    cursor.execute(query)
    data = cursor.fetchall()
    columns = [column[0] for column in cursor.description]
    return pd.DataFrame(data, columns=columns)

# Database connection details
host = '127.0.0.1'
user = 'root'
password = 'Rakchi@1808'  # Replace with your actual MySQL password
database = 'subscriber_decline_analysis'  # Replace with your actual database name

# Connect to the database
connection = mysql.connector.connect(host=host, user=user, password=password, database=database)

# Create a cursor object using the connection
cursor = connection.cursor()

# SQL query to fetch subscribers
query = "SELECT * FROM Subscriber;"

# Create a DataFrame from the query result
subscriber_df = create_dataframe_from_query(cursor, query)

# Close the cursor and connection
cursor.close()
connection.close()

# Count the occurrence of each subscription status in the DataFrame
subscription_status_count = subscriber_df['Subscription_status'].value_counts()

# Plot the distribution of subscription status
plt.figure(figsize=(8, 6))
subscription_status_count.plot(kind='bar', color=['green', 'red'])
plt.title('Distribution of Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Database connection details
host = '127.0.0.1'
user = 'root'
password = 'Rakchi@1808'  # Replace with your actual MySQL password
database = 'subscriber_decline_analysis' # Replace with your actual database name

# Connect to the database
connection = mysql.connector.connect(host=host, user=user, password=password, database=database)

# SQL query to find out which content each subscriber watches, with a count
query = """
SELECT
    Subscriber.S_Name,
    COUNT(Content.Content_ID) AS Content_Watched
FROM
    Subscriber
JOIN
    Viewing_History ON Subscriber.Subscriber_ID = Viewing_History.Subscriber_ID
JOIN
    Content ON Viewing_History.Content_ID = Content.Content_ID
GROUP BY
    Subscriber.Subscriber_ID
ORDER BY
    Content_Watched DESC;
"""

# Create a cursor and execute the query
cursor = connection.cursor()
cursor.execute(query)
data = cursor.fetchall()

# Fetch column names from cursor.description for the DataFrame
columns = [column[0] for column in cursor.description]

# Close the cursor and connection
cursor.close()
connection.close()

# Convert the result to a DataFrame
content_watched_df = pd.DataFrame(data, columns=columns)
colors = plt.cm.viridis(np.linspace(0, 1, len(content_watched_df['S_Name'])))
# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(content_watched_df['S_Name'], content_watched_df['Content_Watched'], color=colors)
plt.xlabel('Subscriber Name')
plt.ylabel('Number of Contents Watched')
plt.title('Contents Watched by Each Subscriber')
plt.xticks(rotation=90)
plt.tight_layout()  # Adjust layout to fit the x labels
plt.show()
```
